```python
import pandas as pd
import re  # Regular expressions for cleaning strings

# Load the dataset
animal = pd.read_csv('Animal Dataset.csv')

# List of columns to keep as numerical
numerical_columns = ['Height (cm)', 'Weight (kg)', 'Lifespan (years)', 'Average Speed (km/h)', 
                     'Gestation Period (days)', 'Top Speed (km/h)', 'Offspring per Birth']

# Function to clean and convert ranges to numerical values
def clean_numerical_value(value):
    if pd.isna(value):  # Handle NaN values
        return value
    if isinstance(value, str):
        value = value.strip()  # Remove leading/trailing whitespace
        
        # Remove non-numeric characters (e.g., 'm', 'kg', etc.)
        value = re.sub(r'[^0-9.-]', '', value)  # Keep numbers, dots, and minus signs
        
        if '-' in value:  # Handle ranges like "30-50"
            parts = value.split('-')
            return (float(parts[0]) + float(parts[1])) / 2
        elif 'up to' in value.lower():  # Handle "up to 0.5" (case-insensitive)
            return float(value.lower().replace('up to', '').strip())
        elif '(' in value:  # Handle "2-4 (usually)"
            main_part = value.split('(')[0].strip()
            if '-' in main_part:
                parts = main_part.split('-')
                return (float(parts[0]) + float(parts[1])) / 2
        else:  # Handle other cases (e.g., single values)
            try:
                return float(value)  # Try converting to float
            except ValueError:
                return None  # If conversion fails, return None
    return value  # If already a number, return as is

# Apply the cleaning function to the numerical columns
for column in numerical_columns:
    animal[column] = animal[column].apply(clean_numerical_value)

# Convert the rest of the columns to categorical
for column in animal.columns:
    if column not in numerical_columns:
        animal[column] = animal[column].astype('category')

# Display the updated DataFrame
print(animal.dtypes)
```

### Why to Standardize?
- Standardize data is easier to visualize because all variables are on the same scale. For example, you can create heatmaps or pair plots without worring about about one variable overshadowing others.
- Standardization ensures that no single variable dominates the analysis due to its scale.
- It improves the performance of algorithms like regression, clustering and PCA
- It makes it easier to intepret and compare the effects of different variables

### When Not to Standardize?
- Categorical Variables: Variables like Animal, Color, Diet, Habitat, etc., are categorical and should not be standardized. Instead, they should be encoded (e.g. using one-hot encoding).
- Count Data: Variables like Offspring per Birth are discrete counts and might not benefit from standardization. Intead, consider using a Poisson distribution or log transformation.

### How Standarization Works?
Standarization transforms the data so the data has: 1) A mean of 0. 2) A standard deviation of 1.

```python
# Function to apply grouping
def group_categories(column, grouping_dict):
    # Step1: store original categories
    original_categories = set(animal[column].cat.categories)
    
    # Step2: Add new categories to the column
    # grouping_dict.keys()：获取 grouping_dict 字典的所有键（即分组后的新类别，如 'Grasslands', 'Forests', 'Aquatic' 等）
    # animal[column].cat.categories：获取 animal[column] 列当前的分类（Categorical）类型中的所有类别
    # cat.categories in Pandas returns all the unique categories defined for a categorical column
    new_categories = set(grouping_dict.keys()) - set(animal[column].cat.categories)
    # animal[column] 列是 Categorical 类型，只能包含预定义的类别。如果直接尝试将一个新的类别赋值给该列（例如 animal[column] = 'Aquatic'），
    # 而 'Aquatic' 不是该列已有的类别，就会抛出 TypeError。
    if new_categories:
        animal[column] = animal[column].cat.add_categories(new_categories)
    
    # Step3: Group categories
    for group, categories in grouping_dict.items():
        # loc: 是pandas中用于给予标签选择数据的方法
        # loc的一滴一个参数是row选择条件(df[column].isin(categories)), 第二个参数是column的选择
        # 整体意思是L选择df中 df[column].isin(categories) 为True 的行， 第二个参数是列选择column这一列
        animal.loc[animal[column].isin(categories), column] = group
    
    # Step4: Remove unused categories
    animal[column] = animal[column].cat.remove_unused_categories()
    
    # Step5: Find removed categories
    final_categories = set(animal[column].cat.categories)
    removed_categories = original_categories - final_categories
    
    # Step6: Print removed categories as a list
    if removed_categories:
        print(f"Removed categories in column '{column}': ")
        print(list(removed_categories)) # Convert set to list for better readability
    # else:
    #     print(f"No categories were removed in column '{column}'.")
        
# Apply grouping to Diet and Conservation Status
group_categories('Diet', diet_grouping)
group_categories('Habitat', habitat_grouping)
group_categories('Conservation Status', conservation_grouping)
```

***2. Convert Ranges to Averages (Already Done)***
- If a column contains ranges like "50-100", we extract and compute the average:

***Why convert ranges to averages?***
- For numerical analysis & modeling (e.g., regression, correlation, clustering, ML models), numerical values are required.
- Statistical summaries (e.g., mean, standard deviation, distribution analysis) work best with continuous values.
- Comparability: Helps in comparing different animals based on characteristics like height, weight, lifespan, etc.

```python
# part A: Function to convert ranges to averages
def range_to_average(value):
    if isinstance(value, str): # Check if it's a string (range)
        try:
            low, high = map(int, value.split('-'))
            return (low + high) / 2
        except ValueError: # handle cases like "up to 0.5"
            try:
                return float(value.split(" ")[-1]) # Take the last number
            except:
                return value # if all fails, return the original string
    return value # Return as is if not a range
```

```python
# part B: Apply the conversion and calculate averages
for column in animal.columns:
    animal[column] = animal[column].apply(range_to_average)

int_cols = ['Height (cm)', 'Weight (kg)', 'Lifespan (years)', 'Gestation Period (days)', 'Top Speed (km/h)', 'Offspring per Birth']

for col in int_cols:
    try:
        animal[col] = animal[col].astype('int')
    except:
        pass
    
animal
```
