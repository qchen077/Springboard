## Random Forest

Random Forest is an ensemble of Decision Trees. With a few exceptions, a `RandomForestClassifier` has all the hyperparameters of a `DecisionTreeClassifier` (to control how trees are grown), plus all the hyperparameters of a `BaggingClassifier` to control the ensemble itself.

The Random Forest algorithm introduces extra randomness when growing trees; instead of searching for the very best feature when splitting a node, it searches for the best feature among a random subset of features. This results in a greater tree diversity, which (once again) trades a higher bias for a lower variance, generally yielding an overall better model. The following `BaggingClassifier` is roughly equivalent to the previous `RandomForestClassifier`. Run the cell below to visualize a single estimator from a random forest model, using the Iris dataset to classify the data into the appropriate species.

```python
from sklearn.tree import export_graphviz
from graphviz import Source
from sklearn.datasets import load_iris
iris = load_iris()

# Model (can also use single decision tree)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10)

# Train
model.fit(iris.data, iris.target)
# Extract single tree
estimator = model.estimators_[5]

# Export to DOT format and display inline
dot_data = export_graphviz(
    estimator,
    out_file=None,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    rounded=True,
    filled=True
)

# Export as dot file
export_graphviz(estimator, out_file='tree.dot', 
                feature_names = iris.feature_names,
                class_names = iris.target_names,
                rounded = True, proportion = False, 
                precision = 2, filled = True)

# Convert to png using system command (requires Graphviz)
from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=600'])

# Display in jupyter notebook
from IPython.display import Image
Image(filename = 'tree.png')
```

Notice how each split seperates the data into buckets of similar observations. This is a single tree and a relatively simple classification dataset, but the same method is used in a more complex dataset with greater depth to the trees.

```python
# RandomForestClassifier — an ensemble learning method using many decision trees and bagging to improve accuracy and prevent overfitting.
from sklearn.ensemble import RandomForestClassifier

# n_estimators=300: Builds 300 decision trees in the forest.
# random_state=1: Ensures reproducibility (you get the same results every run).
# n_jobs=-1: Uses all CPU cores available for parallel processing (faster training).
clf = RandomForestClassifier(n_estimators=300, random_state = 1,n_jobs=-1)
model_res = clf.fit(X_train_scaled, y_train)

y_pred = model_res.predict(X_test_scaled)
# Predicts class probabilities instead of hard labels.
y_pred_prob = model_res.predict_proba(X_test_scaled)
# Extracts the probability of the positive class (class 1), often used for ROC AUC or threshold tuning.
lr_probs = y_pred_prob[:,1]

ac = accuracy_score(y_test, y_pred)

# average='weighted': Accounts for class imbalance.
# Disease as response variable, this is a binary classification (0 = not diseased, 1 = diseased). In public health datasets, cases vs non-cases often look like 
# 90% not disease and 10% diseased.
# Other options available for binary classification:
    # average='binary' (only works for binary)
    # average='macro' (treats all classes equally, regardless of size)
    # average='micro' (averages globally)
f1 = f1_score(y_test, y_pred, average='weighted')
cm = confusion_matrix(y_test, y_pred)

print('Random Forest: Accuracy=%.3f' % (ac))

print('Random Forest: f1-score=%.3f' % (f1))
```

### Why are the correlations so low, but the classifier's accuracy is high?

You're seeing a classic and interesting phenomenon in machine learning:

***1. Correlation only captures linear relationships:***

    . The correlation heatmap (like Pearson correlation) only shows linear associations between two variables.
    . If a variable has a nonlinear or conditional relationship with the target, correlation will still appear low.

***✅ But Random Forest is a nonlinear model — it can detect:***

    . If-else style splits
    . Interactions between variables
    . Threshold-based rules that correlation can't capture

So even if correlations are near 0, the model can still learn powerful decision rules from combinations of features.

***2. Multiple weak signals can create a strong signal:***

    . Each feature may be weakly related to the target on its own (low correlation).
    . But together, they can form patterns that are very predictive.

***For example:***
    Age + infection_case + sex might be weak individually, but combined: "Female over 60 who visited Shincheonji Church" might have high disease risk.

Random Forest can build this logic automatically across trees.

***3. You're using dummy variables (one-hot encoded categories)***

    . You converted categorical variables (like infection_case, state, sex) into multiple binary (0/1) features.
    . These aren’t shown in the numeric-only heatmap.
    . But in classification tasks, these dummy variables are often extremely informative.

Even though correlation doesn’t show them, the model uses them effectively.

### Create Confusion Matrix Plots
Confusion matrices are great ways to review your model performance for a multi-class classification problem. Being able to identify which class the misclassified observations end up in is a great way to determine if you need to build additional features to improve your overall model. In the example below we plot a regular counts confusion matrix as well as a weighted percent confusion matrix. The percent confusion matrix is particulary helpful when you have unbalanced class sizes.

```python
class_names=['isolated','released','missing','deceased'] # name  of classes
```

```python
# itertools.product: lets you loop over all (i, j) index pairs in the matrix
import itertools
import numpy as np
import matplotlib.pyplot as plt

from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

# cm: The confusion matrix (2D array)
# cmap=plt.cm.Blues: built-in colormap in Matplotlib. (option: cmap=plt.cm.viridis)
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.viridis):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    # If normalization is on, convert counts to percentages (row-wise) (each row will sum to 1).
    if normalize:
        # cm.sum(axis=1) → Returns a 1D array of row sums (shape: (n,))
        # [:, np.newaxis] → Converts that array to a column vector (shape: (n, 1)), so it can be used for broadcasting division.
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    # Shows the matrix as a heatmap (blues by default)
    # plt.imshow() is used to display an image (in this case, your confusion matrix as a heatmap).
    # interpolation='nearest' means:
        # Don’t smooth the image — display each cell as a solid block of color.
        # This keeps the boxes crisp and square — perfect for a grid-like structure like a confusion matrix.
    # Other interpolation modes ('bilinear' or 'bicubic') make sense for photos or continuous data, but not for categorical grids like confusion matrices.
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    # range(cm.shape[0]): Generates values 0 to cm.shape[0] - 1 → to loop over rows
    # range(cm.shape[1]): Generates values 0 to cm.shape[1] - 1 → to loop over columns
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        # plt.text(j, i, format(cm[i, j], fmt), ...): Adds the actual number/percentage inside each cell
        # imshow() uses (x, y) = (column, row) coordinates, so plt.text(j, i, ...)
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()


# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Confusion matrix, without normalization')
#plt.savefig('figures/RF_cm_multi_class.png')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')
#plt.savefig('figures/RF_cm_proportion_multi_class.png', bbox_inches="tight")
plt.show()
```

### Plot feature importances
The random forest algorithm can be used as a regression or classification model. In either case it tends to be a bit of a black box, where understanding what's happening under the hood can be difficult. Plotting the feature importances is one way that you can gain a perspective on which features are driving the model predictions.

```python
# What are the features didn't make into the feature importances? but why all indexs, I need text labels.
feature_importance = clf.feature_importances_
top_30_idx = np.argsort(feature_importance)[-30:]
top_30_idx
```

```python
# Gets the importance score for each feature from the trained model (clf)
feature_importance = clf.feature_importances_

# make importances relative to max importance
# Measures how much each feature reduces impurity (like Gini or entropy) across all trees.
# So the top feature will always have a value of 100
feature_importance = 100.0 * (feature_importance / feature_importance.max())[:30]

# Sorts the top 30 feature importances in ascending order (for a better-looking horizontal bar chart)
sorted_idx = np.argsort(feature_importance)[:30]

# Creates evenly spaced positions for plotting bars on the y-axis.
pos = np.arange(sorted_idx.shape[0]) + .5
# print(pos.size)

# Create color mapping using viridis
colors = plt.cm.viridis(feature_importance[sorted_idx] / 100.0)

sorted_idx.size
plt.figure(figsize=(10,10))
plt.barh(pos, feature_importance[sorted_idx], color=colors, align='center')
plt.yticks(pos, X.columns[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')

plt.show()
```

### What's next? How to improve this decision tree's prediction performance?

***1. Feature Engineering***

    - Add new features: Create features that capture domain logic. E.g., ratios, interactions, or flags.
    - Bin continuous features: Sometimes bucketing numerical variables can help decision trees capture patterns better.
    - Handle rare categories: Merge or drop very rare values in categorical columns to reduce noise.
    - Date/time features: If you have timestamps, extract day-of-week, hour, etc.
    
***2. Hyperparameter Tuning***

    A vanilla decision tree is rarely optimal. Use tools like GridSearchCV or RandomizedSearchCV to tune parameters 
    
***3. Try Ensemble Models***

    Decision Trees are weak learners by nature. You’ll see dramatic improvement by using:
        ✅ Random Forest – reduces overfitting and improves accuracy.
        ✅ Gradient Boosting (e.g., XGBoost, LightGBM, CatBoost) – powerful for tabular data with better prediction performance.

***4. Evaluate the Model Properly***

    Make sure you're using the right metrics for regression or classification. For regression:
        - MAE, MSE, RMSE
        - R² score
        - Visual inspection of residuals

    For classification:
        - Accuracy, precision, recall, F1
        - ROC-AUC
        - Confusion matrix
